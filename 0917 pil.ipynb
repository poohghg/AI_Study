```python
from PIL import Image
import numpy as np
```

```python
im = Image.open("demo1.jpg")
```

```python
# (left, upper, right, lower)-tuple.
im.crop((10,10,100,100))
```

```python
im.rotate(30)
```

```python
# filow 와 numpy의 연동
x = np.array(im)
```

```python
# 어레이로부터 이미지로 처리
Image.fromarray(x)
```
