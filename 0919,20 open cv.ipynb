```python
# 255 벨류를 가진 하얀색 도화지를 생성
img3 = np.full((512,512,3),255,np.uint8)
# line(img, pt1, pt2, color[, thickness[, lineType[, shift]]]) -> img
# 색깔을 4개로 표현하면 그림자 까지 표현한다.
img3 = cv2.line(img3,(0,0),(511,511),(0,0,255),5)
plt.imshow(img3)
```

```python
img4 = np.zeros((512,512,3),np.uint8)
img4 = cv2.rectangle(img4,(100,0),(510,128),(0,255,0),5)
plt.imshow(img4)
```

```python
# 다각형 그리기
img6 = np.zeros((512,512,3),np.uint8)
pts = np.array([[10,5],[20,30],[70,20],[50,10]],np.int32)
# 이미지에 표현하기 위해 3차원 행렬로 변환, 변환이전과 이후의 행렬 갯수는 동일 해야함
# 
pts = pts.reshape((-1,1,2))
img6 = cv2.polylines(img6,[pts],True,(0,255,255))
plt.imshow(img6)
```
