```python
from skimage.measure import compare_ssim
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import imutils
```

```python
imageA = cv2.imread('assets/similarity/1.jpg', 1)
imageB = cv2.imread('assets/similarity/2.png', -1)
```

```python
plt.imshow(imageB)
```

```python
 grayA = cv2.cvtColor(imageA, cv2.COLOR_BGR2GRAY)
grayB = cv2.cvtColor(imageB, cv2.COLOR_BGR2GRAY)
```

```python
compare_ssim(grayA, grayB, full=True)
```

```python
(score, diff) = compare_ssim(grayA, grayB, full=True)
diff = (diff * 255).astype("uint8")
score
```

```python
thresh = cv2.threshold(diff, 0, 255,
	cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]
cnts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL,
	cv2.CHAIN_APPROX_SIMPLE)
```

```python
cnts = imutils.grab_contours(cnts)
```

```python
for c in cnts:
	(x, y, w, h) = cv2.boundingRect(c)
	cv2.rectangle(imageA, (x, y), (x + w, y + h), (0, 0, 255), 2)
	cv2.rectangle(imageB, (x, y), (x + w, y + h), (0, 0, 255), 2)
```

```python
fig, axes = plt.subplots(2,2, True, True, figsize=(15,15))
axes[0,0].imshow(imageA)
axes[0,1].imshow(imageB)
axes[1,0].imshow(diff)
axes[1,1].imshow(thresh)
plt.tight_layout()
```
