## create_model()

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# 정규화 모델 생성
mm = MinMaxScaler()
```

```python
# 데이터 정규화
iris_norm = mm.fit_transform(iris.data)
```

```python
X_train,X_test,y_train,y_test = train_test_split(iris_norm,iris.target) 
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(16, activation='relu', input_shape=(4,)))
    model.add(Dense(16, activation='relu'))
    model.add(Dense(3, activation='softmax'))
    model.compile(loss='sparse_categorical_crossentropy',
                 optimizer='adam', metrics=['acc'])
    return model
```

```python
kc = KerasClassifier(create_model,epochs=50)
```

```python
kc.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
kc2 = KerasClassifier(create_model,epochs=50) # epochs 넣을 수 있음. sk_params 속지않음
```

```python
cross_val_score(kc2, X_train,y_train, cv=10) # create_model() 안됨
```
