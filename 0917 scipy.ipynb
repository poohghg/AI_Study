### open 
- 파이썬이 기본적으로 불러올수 있는것은 바이너리와 text
- 파일은 세가지로 구성
- 경로,파일명,확장자
    - 경로는 절대경로와 상대경로가 존재
    - 경로가 없다면 현재 작업공간이다.
    -
- with 는 enter와 exit가 존재할때 사용이 가능하다.

```python
#file,mode = r,w,b
t = open("demo.raw","rb",)
```

```python
dir(t)
```

```python
with open("demo.raw",encoding="utf-8") as t:
    print(t)
```

```python
class T:
    def __enter__(self):
        print('enter')
    def __exit__(self,a,b,c):
        print('exit')
```

```python
t = T()
```

```python
# enter 와 exit가 존재하면 with 구문을 사용할 수 있다.
# 이는 파이썬이 덕타이핑이기 때문이다.
with T():
    print('aaa')
```

```python
from scipy import misc
```

```python
f = misc.face(gray=True)
plt.figure(figsize=(10,3.6))

# f는 내가 보여주고싶은거
# 넘파이 행 열 차원
plt.subplot(131)
plt.imshow(f[10:50,30:200],cmap=plt.cm.gray)


plt.subplot(132)
plt.imshow(f,cmap=plt.cm.gray,vmin=30,vmax=200)
# 주위 axis를 모두 off
plt.axis("off")

plt.subplot(133)
plt.imshow(f,cmap=plt.cm.gray)

# 윤곽선을 나타냄
plt.contour(f,[50,200])
plt.axis("off")

# 옵션을 줘서 공간의 위츠를 바꿀수 있다.
#plt.subplots_adjust(wspace=0,hspace=0,top=0.99,bottom=0.01,right=0.99)

# tight _layout 공간을 적절히 분배해줌
plt.tight_layout()
```
